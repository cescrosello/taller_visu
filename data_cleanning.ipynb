```python
import matplotlib.pyplot as plt

# 1. Distribución de la Edad de los Clientes
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black', )
plt.title('Distribución de la Edad de los Clientes')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
# 2. Cantidad de Compras por Categoría
plt.figure(figsize=(10, 6))
df['category'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Cantidad de Compras por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Cantidad de Compras')
plt.grid(True)
plt.show()
```

```python
# 3. Método de Pago Utilizado
plt.figure(figsize=(10, 6))
df['payment_method'].value_counts().plot(kind='bar', color='salmon')
plt.title('Método de Pago Utilizado')
plt.xlabel('Método de Pago')
plt.ylabel('Cantidad de Compras')
plt.grid(True)
plt.show()
```

```python
# 4. Género de los Clientes
plt.figure(figsize=(8, 8))
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Género de los Clientes')
plt.ylabel('')
plt.show()
```

```python
# 5. Número de Compras por Mes y Año
plt.figure(figsize=(12, 6))
df.groupby(['year', 'month']).size().unstack().T.plot(kind='bar', stacked=True)
plt.title('Número de Compras por Mes y Año')
plt.xlabel('Mes')
plt.ylabel('Cantidad de Compras')
plt.legend(title='Año')
plt.grid(True)
plt.show()
```

```python
# 6. Precio Total por Categoría
plt.figure(figsize=(10, 6))
df.groupby('category')['price'].sum().plot(kind='bar', color='lightseagreen')
plt.title('Precio Total por Categoría')
plt.xlabel('Categoría')
plt.ylabel('Precio Total')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sumar las ventas por centro comercial y año
sales_by_mall_year = df.groupby(['shopping_mall', 'year'])['price'].sum().reset_index()

# Crear la gráfica de barras
plt.figure(figsize=(12, 6))
sns.barplot(data=sales_by_mall_year, x='shopping_mall', y='price', hue='year')
plt.title('Ventas por Año en Cada Centro Comercial')
plt.xlabel('Centro Comercial')
plt.ylabel('Ventas (en unidades monetarias)')
plt.legend(title='Año')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Contar el número de transacciones por centro comercial y género
transactions_by_mall_gender = df.groupby(['shopping_mall', 'gender'])['invoice_no'].count().reset_index()

# Renombrar la columna de conteo para claridad
transactions_by_mall_gender.rename(columns={'invoice_no': 'transactions'}, inplace=True)

# Crear la gráfica de barras
plt.figure(figsize=(12, 6))
sns.barplot(data=transactions_by_mall_gender, x='shopping_mall', y='transactions', hue='gender')
plt.title('Número de Transacciones por Género en Cada Centro Comercial')
plt.xlabel('Centro Comercial')
plt.ylabel('Número de Transacciones')
plt.legend(title='Género')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Sumar el gasto total por centro comercial y género
spending_by_mall_gender = df.groupby(['shopping_mall', 'gender'])['price'].sum().reset_index()

# Crear la gráfica de barras
plt.figure(figsize=(12, 6))
sns.barplot(data=spending_by_mall_gender, x='shopping_mall', y='price', hue='gender')
plt.title('Gasto Total por Género en Cada Centro Comercial')
plt.xlabel('Centro Comercial')
plt.ylabel('Gasto Total (en unidades monetarias)')
plt.legend(title='Género')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
